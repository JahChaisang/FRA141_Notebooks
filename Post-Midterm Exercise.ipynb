1. เขียนฟังก์ชันชื่อ sum_list ที่รับค่าเป็นลิสต์ ชื่อ lst แล้วรีเทิร์นผลลัพธ์เป็นผลบวกของตัวเลขภายในลิสต์นั้น เช่น
```
>>> sum_list([3,4,2,1])
10
>>> sum_list([8,8,4])
20
```

2. เขียนฟังก์ชันชื่อ len_square ที่รับค่าเป็นลิสต์แล้วรีเทิร์นผลลัพธ์เป็นจำนวนสมาชิกของลิสต์ยกกำลังสอง
```
>>> len_square([3,4,2,1])
16
>>> len_square([8,8,4])
9
```

3. เขียนฟังก์ชันชื่อ count_upper ที่รับค่าเป็นสตริงแล้วรีเทิร์นผลลัพธ์เป็นจำนวนตัวอักษรที่เป็นตัวพิมพ์ใหญ่ เช่น
```
>>> count_upper('AAjdalsN')
3
>>> len_square('PPAPisagoodSoNg')
6
```

4. เขียนฟังก์ชันชื่อ div_all_by_2 ที่รับค่าเป็นลิสต์แล้วรีเทิร์นผลลัพธ์เป็นค่าของเลขในลิสต์นั้นหารสอง เช่น
```
>>> div_all_by_2([3,4,2,1])
[1.5 2 1 0.5]
>>> div_all_by_2([8,8,4])
[4 4 2]
```

5. เขียนฟังก์ชันชื่อ count_str ที่รับค่าเป็นลิสต์แล้วรีเทิร์นผลลัพธ์เป็นจำนวนสมาชิกในลิสต์ที่เป็นตัวอักษร
```
>>> count_str(["a",6,4,"b"])
2
>>> count_str(["cute","nice",45,32,222,"cd","okay"])
4
```

6. เขียนฟังก์ชันชื่อ upper_even ที่รับค่าเป็นสตริงแล้วเปลี่ยนตัวอักษรตำแหน่งที่เป็นเลขคู่่เช่น 0,2,4,6,... เป็นตัวพิมพ์ใหญ่ ที่เหลือเป็นตัวพิมพ์เล็ก
```
>>> upper_even('aadjskJJl')
AaDjSkJjL
>>> upper_even('helloworld')
HeLlOwOrLd
```

7. เขียนฟังก์ชันชื่อ gen_initial ที่รับค่าเป็นสตริงซึ่งเป็นชื่อและนามสกุลของคน ทั้งชื่อและนามสกุลจะเริ่มด้วยตัวพิมพ์ใหญ่ตามด้วยตัวพิมพ์เล็ก รีเทิร์นผลลัพธ์เป็นชื่อย่อของคนคนนั้น เช่น
```
>>> gen_initial("John Doe")
JD
>>> gen_initial("Peter Griffin")
PG
```

10. อาจารย์สร้างดิกชันนารีขึ้นมาหนึ่งอันเพื่อเก็บคะแนนสอบของนักศึกษาดังนี้
```
students = {'Ben': 25, 'Lisa': 40, 'Aron': 28}
```
ต่อมาอาจารย์ต้องการเปลี่ยนแปลงคะแนนของนักศึกษาบางคน เขียนฟังก์ชันชื่อ add_score ที่รับตัวแปรสามตัวคือ ดิกชันนารีชื่อ students, ชื่อนักศึกษา, และคะแนนที่อยากจะเพิ่ม แล้วรีเทิร์นค่าเป็นดิกชันนารีที่เปลี่ยนแปลงแก้ไขแล้ว เช่น
```
>>> add_score(students, 'Ben', 4)
{'Ben': 29, 'Lisa': 40, 'Aron': 28}
```

9. อาจารย์สร้างดิกชันนารีขึ้นมาหนึ่งอันเพื่อเก็บคะแนนสอบสองครั้งของนักศึกษาดังนี้
```
students = {'Ben': [25, 50], 'Lisa': [40, 34], 'Aron': [28, 33]}
```
ต่อมาเมื่อมีการสอบอีกหนึ่งครั้งอาจารย์ก็ต้องการที่จะใส่คะแนนเข้าไปเพิ่มให้นศ.แต่ละคน เขียนฟังก์ชันชื่อ append_score ที่รับตัวแปรสองตัวคือ ดิกชันนารีชื่อ students ชื่อนักศึกษา และคะแนนสอบครั้งใหม่ แล้วรีเทิร์นผลลัพธ์เป็นดิกชันนารีใหม่ หลังจากเติมคะแนนเข้าไปในลิสต์ของนศ.คนนั้นแล้วเช่น
```
>>> append_score(students, 'Lisa', 17)
students = {'Ben': [25, 50], 'Lisa': [40, 34, 17], 'Aron': [28, 33]}
```


10. อาจารย์สร้างดิกชันนารีขึ้นมาหนึ่งอันเพื่อเก็บคะแนนสอบสามครั้งของนักศึกษาดังนี้
```
students = {'Ben': [25, 50, 17], 'Lisa': [40, 34, 12], 'Aron': [28,33,18]}
```
เขียนฟังก์ชันชื่อ average_score ที่รับตัวแปรสองตัวคือ ดิกชันนารีชื่อ students และชื่อนักศึกษา แล้วรีเทิร์นผลลัพธ์เป็นค่าคะแนนเฉลี่ยของนศ.คนนั้น เช่น
```
>>> average_score(students, 'Lisa')
28.67
```
