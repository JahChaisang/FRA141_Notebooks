## แบบฝึกหัดสำหรับ Quiz 2


**ข้อ 1**
ถ้าวันนั้นเป็นวันเสาร์อาทิตย์และสภาพอากาศเหมาะสม เราจะออกไปเที่ยว เขียนโปรแกรมที่เช็คว่าวันนี้เราจะไปเที่ยวหรือไม่ เช่นเขียนฟังก์ชันชื่อ travel_check ที่รับสองตัวแปรคือ temperature และ weekend ถ้าอุณหภูมิอยู่ระหว่าง 25-35 องศาและ weekend มีค่าเป็น True ให้รีเทิร์น True (เราไปเที่ยว) มิฉะนั้นให้รีเทิร์น False (เราไม่ไปเที่ยว)
```
>>> travel_check(33, True)
True
>>> travel_check(33, False)
False
>>> travel_check(38, False)
False
>>> travel_check(40, True)
False
```

**ข้อ 2**

เขียนฟังก์ชันชื่อ sum_maybe ที่รับค่าเป็นจำนวนเต็มสองค่า (a และ b) ฟังก์ชันนี้ควรรีเทิร์นผลบวกของ a และ b แต่ถ้าผลบวกของ a และ b อยู่ระหว่าง 10 ถึง 19 ให้รีเทิร์น 20
```
>>> sum_maybe(3, 4)
7
>>> sum_maybe(9, 4)
20
>>> sum_maybe(10, 11)
21
```

**ข้อ 3**

เขียนโปรแกรมตั้งนาฬิกาปลุก สมมติว่าวันแต่ละวันจะมีโค้ด เช่น วันอาทิตย์=1, วันจันทร์=2, วันเสาร์=6 เขียนฟังก์ชัน set_alarm ให้รับค่าจำนวนเต็ม day_code ถ้าเป็นวันธรรมดา (day_code คือ 2 ถึง 5) เราต้องตื่นเช้าให้รีเทิร์น ‘7:00’ ถ้าเป็นวันเสาร์อาทิตย์ (day_code คือ 1 หรือ 6) ให้รีเทิร์น ’10:00’
```
>>> set_alarm(1)
’10:00’
>>> set_alarm(5)
’7:00’
```

**ข้อ 4**

เขียนฟังก์ชันชื่อ check6 ที่รับตัวแปรเป็นจำนวนเต็มสองค่า (a และ b) ฟังก์ชันนี้จะรีเทิร์นค่า True ในสามกรณี (1) a หรือ b มีค่าเท่ากับ 6, (2) ผลบวกของสองตัวเลขเท่ากับ 6, (3) ผลต่างของสองตัวเลขเท่ากับ 6 กรณีอื่นฟังก์ชันนี้จะรีเทิร์นค่า False
```
>>> check6(6, 4)
True
>>> check6(4, 5)
False
>>> check6(1, 5)
True
>>> check6(12, 6)
True
```

**ข้อ 5**

เขียนฟังก์ชันชื่อ is1to10 ที่รับตัวแปรสองตัวเป็นจำนวนเต็ม (n) และ boolean (mode) ในโหมดแรก (mode=1) ฟังก์ชันนี้จะรีเทิร์นค่า True ต่อเมื่อ n มากกว่า 1 และน้อยกว่า 10 ในโหมดที่สอง (mode=2) ฟังก์ชันนี้จะรีเทิร์นค่า True ต่อเมื่อ n น้อยกว่าหรือเท่ากับ 1 หรือ n มากกว่าหรือเท่ากับ 10

```
>>> is1to10(5, 1)
True
>>> is1to10(11, 2)
True
>>> is1to10(11, 1)
False
>>> is1to10(1, 2)
True
>>> is1to10(1, 1)
False
```
